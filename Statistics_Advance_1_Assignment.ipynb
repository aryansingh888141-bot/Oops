```python
#Q-9. Write a Python program to generate 1000 random number from a normal distribution with mean = 5. Compute its mean and
#      standard deviation using Numpy. and draw a histogram to visualize the disribution.
data = np.random.normal(5, 1, 1000)
mean = np.mean(data)
std = np.std(data)
print("Mean:", mean)
print("Standard Deviation:", std)
plt.hist(data, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution")
plt.show()
```

```python
#Q-10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and
#       wants you to indentify the overall sales trend.
#     daily_sales = [220,245,210,265,230,250,260,275,240,255,235,260,245,250,225,270,265,255,250,260]
#    * Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
#    * Write the Python code to compute the mean sales and its confidence interval.
daily_sales = [220,245,210,265,230,250,260,275,240,255,235,260,245,250,225,270,265,255,250,260]
import numpy as np
import scipy.stats as stats
mean = np.mean(daily_sales)
std = np.std(daily_sales)
n = len(daily_sales)
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * (std / np.sqrt(n))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print("Mean sales:", mean)
print("Confidence interval:", confidence_interval)
```
